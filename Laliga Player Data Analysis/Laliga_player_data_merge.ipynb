```python
import pandas as pd 
```

```python
df1=pd.read_csv("Laliga Player Scores.csv")
df2=pd.read_csv("Laliga Player Assists.csv")
df3=pd.read_csv("Laliga Player Passes.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
df1.sort_values(['TEAM','NAME'],inplace=True)
df2.sort_values(['TEAM','NAME'],inplace=True)
df3.sort_values(['TEAM','NAME'],inplace=True)
```

```python
df1.reset_index(inplace=True)
df1.drop('index',axis=1,inplace=True)
```

```python
df1.head()
```

```python
df2.reset_index(inplace=True)
df2.drop('index',axis=1,inplace=True)
df2.head()
```

```python
df3.reset_index(inplace=True)
df3.drop('index',axis=1,inplace=True)
df3.head()
```

```python
laliga_stat_df=pd.merge(pd.merge(df1, df2, on=['TEAM', 'NAME','GAMES'], how='inner'), df3, on=['TEAM', 'NAME'], how='inner')
```
